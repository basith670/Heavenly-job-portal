```python
file_storage_service_java_code = """
package com.heavenly.util;
import org.springframework.stereotype.Service;
import org.springframework.web.multipart.MultipartFile;
import java.nio.file.*;
import java.io.IOException;
import java.util.UUID;

@Service
public class FileStorageService {
  private final Path root = Paths.get("uploads");
  public FileStorageService() throws IOException { Files.createDirectories(root); }
  public String store(MultipartFile file) throws IOException {
    String ext = "";
    String name = file.getOriginalFilename();
    int i = name.lastIndexOf('.');
    if(i>0) ext = name.substring(i);
    String fname = UUID.randomUUID().toString()+ext;
    Path dest = root.resolve(fname);
    Files.copy(file.getInputStream(), dest, StandardCopyOption.REPLACE_EXISTING);
    return dest.toString();
  }
  public Path load(String path) { return Paths.get(path); }
}
"""
```

```python
auth_controller_java_code = """
package com.heavenly.controller;
import org.springframework.web.bind.annotation.*;
import org.springframework.http.ResponseEntity;
import org.springframework.security.crypto.bcrypt.BCryptPasswordEncoder;
import org.springframework.beans.factory.annotation.Autowired;
import com.heavenly.repo.UserRepo;
import com.heavenly.model.User;
import java.util.Map;

@RestController
@RequestMapping("/api/auth")
public class AuthController {
  @Autowired UserRepo userRepo;
  @Autowired BCryptPasswordEncoder encoder;

  @PostMapping("/register")
  public ResponseEntity<?> register(@RequestBody Map<String,String> body){
    if(userRepo.findByEmail(body.get("email")).isPresent()) {
      return ResponseEntity.badRequest().body("Email exists");
    }
    User u = new User();
    u.setName(body.get("name"));
    u.setEmail(body.get("email"));
    u.setRole(body.getOrDefault("role","SEEKER"));
    u.setPasswordHash(encoder.encode(body.get("password")));
    u.setSkills(body.get("skills"));
    u.setLocation(body.get("location"));
    userRepo.save(u);
    u.setPasswordHash(null);
    return ResponseEntity.ok(u);
  }

  @PostMapping("/login")
  public ResponseEntity<?> login(@RequestBody Map<String,String> body){
    return userRepo.findByEmail(body.get("email"))
      .map(u -> {
        if(encoder.matches(body.get("password"), u.getPasswordHash())){
          u.setPasswordHash(null);
          return ResponseEntity.ok(u);
        } else return ResponseEntity.status(401).body("Bad credentials");
      }).orElse(ResponseEntity.status(404).body("User not found"));
  }
}
"""
```

```python
job_controller_java_code = """
package com.heavenly.controller;
import org.springframework.web.bind.annotation.*;
import org.springframework.beans.factory.annotation.Autowired;
import com.heavenly.repo.JobRepo;
import com.heavenly.model.Job;
import java.util.List;

@RestController
@RequestMapping("/api/jobs")
public class JobController {
  @Autowired JobRepo repo;

  @PostMapping
  public Job postJob(@RequestBody Job job){ return repo.save(job); }

  @GetMapping
  public List<Job> allJobs(){ return repo.findAll(); }

  @GetMapping("/search")
  public List<Job> search(@RequestParam(required=false) String q, @RequestParam(required=false) String location) {
    // simple search: filter in Java (demo). For real app create queries.
    return repo.findAll().stream()
      .filter(j -> (q==null || j.getTitle().toLowerCase().contains(q.toLowerCase()) || j.getDescription().toLowerCase().contains(q.toLowerCase()))
                && (location==null || j.getLocation().toLowerCase().contains(location.toLowerCase())))
      .toList();
  }
}
"""
```

```python
application_controller_java_code = """
package com.heavenly.controller;
import org.springframework.web.bind.annotation.*;
import org.springframework.beans.factory.annotation.Autowired;
import org.springframework.web.multipart.MultipartFile;
import com.heavenly.repo.*;
import com.heavenly.model.*;
import com.heavenly.util.FileStorageService;
import java.time.Instant;

@RestController
@RequestMapping("/api/applications")
public class ApplicationController {
  @Autowired JobApplicationRepo appRepo;
  @Autowired FileStorageService files;

  @PostMapping("/apply")
  public JobApplication apply(@RequestParam Long jobId, @RequestParam Long userId, @RequestParam MultipartFile resume) throws Exception {
    String path = files.store(resume);
    JobApplication a = new JobApplication();
    a.setJobId(jobId); a.setUserId(userId);
    a.setResumePath(path); a.setStatus("APPLIED");
    a.setAppliedAt(Instant.now());
    return appRepo.save(a);
  }

  @GetMapping("/byJob/{jobId}")
  public java.util.List<JobApplication> appsForJob(@PathVariable Long jobId){
    return appRepo.findAll().stream().filter(a->a.getJobId().equals(jobId)).toList();
  }
}
"""
```

## Translate Job Controller

### Subtask:
Translate the `JobController.java` into a Python Flask blueprint/routes, including the necessary repository.


**Reasoning**:
First, I will create the `JobRepository` class with `save`, `find_all`, and `search` methods, mirroring the functionality of `JobRepo.java`, and then save it to `job_repo.py`.



```python
job_repo_py_code = """
from app import db
from job_model import Job

class JobRepository:
    @staticmethod
    def save(job):
        db.session.add(job)
        db.session.commit()
        return job

    @staticmethod
    def find_all():
        return Job.query.all()

    @staticmethod
    def search(query, location):
        jobs_query = Job.query

        if query:
            search_term = f'%{{query}}%'
            jobs_query = jobs_query.filter(
                (Job.title.ilike(search_term)) |
                (Job.description.ilike(search_term))
            )

        if location:
            location_term = f'%{{location}}%'
            jobs_query = jobs_query.filter(Job.location.ilike(location_term))

        return jobs_query.all()
"""

# To save this to a file:
with open('job_repo.py', 'w') as f:
    f.write(job_repo_py_code)

print("Generated job_repo.py code for the Job SQLAlchemy repository.")
```

**Reasoning**:
Now I will define the `job_controller_py_code` string with the Flask blueprint and its routes as specified in the instructions, handling JSON parsing and serialization, and then save it to `job_controller.py`.



```python
job_controller_py_code = """
from flask import Blueprint, request, jsonify
from job_model import Job
from job_repo import JobRepository

job_bp = Blueprint('job_bp', __name__, url_prefix='/api/jobs')

def job_to_dict(job):
    # Helper to convert Job object to a dictionary for JSON serialization
    if job:
        # Exclude internal SQLAlchemy attributes like _sa_instance_state
        job_dict = {k: v for k, v in job.__dict__.items() if not k.startswith('_')}
        return job_dict
    return None

@job_bp.route('/', methods=['POST'])
def post_job():
    data = request.get_json()
    if not data:
        return jsonify({'message': 'Invalid JSON'}), 400

    job = Job(
        title=data.get('title'),
        description=data.get('description'),
        location=data.get('location'),
        jobType=data.get('jobType'),
        postedBy=data.get('postedBy')
    )
    saved_job = JobRepository.save(job)
    return jsonify(job_to_dict(saved_job)), 201

@job_bp.route('/', methods=['GET'])
def get_all_jobs():
    jobs = JobRepository.find_all()
    return jsonify([job_to_dict(job) for job in jobs])

@job_bp.route('/search', methods=['GET'])
def search_jobs():
    q = request.args.get('q')
    location = request.args.get('location')
    jobs = JobRepository.search(q, location)
    return jsonify([job_to_dict(job) for job in jobs])
"""

# To save this to a file:
with open('job_controller.py', 'w') as f:
    f.write(job_controller_py_code)

print("Generated job_controller.py code for the Job Flask blueprint.")
```

## Translate Message Repository

### Subtask:
Translate the `MessageRepo` (implicitly used by `MessageController.java`) into a Python SQLAlchemy repository/DAO.


**Reasoning**:
To translate the `MessageRepo` into a Python SQLAlchemy repository, I will define a Python string variable containing the SQLAlchemy repository code for `message_repo.py`, including necessary imports and methods for database operations, and then save it to a file with a confirmation message.



```python
message_repo_py_code = """
from app import db
from message_model import Message

class MessageRepository:
    @staticmethod
    def save(message):
        db.session.add(message)
        db.session.commit()
        return message

    @staticmethod
    def find_all():
        return Message.query.all()

    @staticmethod
    def find_conversation(user_a_id, user_b_id):
        return Message.query.filter(
            ((Message.fromUser == user_a_id) & (Message.toUser == user_b_id)) |
            ((Message.fromUser == user_b_id) & (Message.toUser == user_a_id))
        ).order_by(Message.sentAt).all()
"""

# To save this to a file:
with open('message_repo.py', 'w') as f:
    f.write(message_repo_py_code)

print("Generated message_repo.py code for the Message SQLAlchemy repository.")
```

```python
security_config_py_code = """
from flask import request, jsonify, g
from user_repo import UserRepository

# Define public paths that do not require authentication
# Matches Java: "/api/auth/**", "/","/index.html","/log.in.html","/assets/**","/h2-console/**"
# A more robust solution might use regular expressions for pattern matching.
PUBLIC_PATHS = [
    '/',
    '/index.html',
    '/log.in.html', # Assuming this is a static HTML file
]
# Prefix-based public paths
PUBLIC_PREFIXES = [
    '/api/auth/',
    '/assets/',
    '/h2-console/', # If you were to run a H2-like console in Flask
]

def configure_security(app):
    @app.before_request
    def check_authentication():
        # Allow OPTIONS requests to pass through for CORS preflight
        if request.method == 'OPTIONS':
            return None # Allow the request to proceed

        path = request.path

        # Check if the current path is a public path
        is_public = path in PUBLIC_PATHS or any(path.startswith(prefix) for prefix in PUBLIC_PREFIXES)

        if is_public:
            return None # Allow public paths to proceed without authentication

        # For all other paths, require HTTP Basic Authentication
        auth = request.authorization
        if not auth or not auth.username or not auth.password:
            return jsonify({'message': 'Authentication required'}), 401, {'WWW-Authenticate': 'Basic realm="Login Required"'}

        user = UserRepository.find_by_email(auth.username)
        if not user or not user.check_password(auth.password):
            return jsonify({'message': 'Invalid credentials'}), 401, {'WWW-Authenticate': 'Basic realm="Login Required"'}

        g.user = user # Store the authenticated user in the global context for the request
        return None # Allow the request to proceed after successful authentication

    # Note on http.headers().frameOptions().disable():
    # In Spring Boot, this is often used to allow the H2 console to be embedded in an iframe.
    # In Flask, such headers are typically managed by the web server (e.g., Nginx, Apache) or a Flask extension
    # like Flask-Talisman. For a direct translation into application code without a specific Flask-based
    # H2 console equivalent, explicitly setting X-Frame-Options is often not required at this level.
    # If it were strictly needed, an @app.after_request handler could set response.headers["X-Frame-Options"] = "ALLOWALL".
"""

# To save this to a file:
with open('security_config.py', 'w') as f:
    f.write(security_config_py_code)

print("Generated security_config.py code for the Flask security configuration.")
```
